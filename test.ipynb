```python
df_players = pd.DataFrame(data["elements"])
df_teams = pd.DataFrame(data["teams"])
df_positions = pd.DataFrame(data["element_types"])
df_gameweeks = pd.DataFrame(data["events"])
```

### Kolom relevan

```python
team_cols = [
    "id", "name", "short_name", "strength",
    "strength_overall_home", "strength_overall_away",
    "strength_attack_home", "strength_attack_away",
    "strength_defence_home", "strength_defence_away",
    "code", "pulse_id"
]

dim_team = df_teams[team_cols]
```

```python
player_cols = [
    "id", "code", "first_name", "second_name", "web_name",
    "element_type", "region", "birth_date", 
    "team", "team_code", "team_join_date", "squad_number", "photo"
]

dim_player = df_players[player_cols]
```

```python
position_cols = [
    "id", "singular_name", "singular_name_short",
    "plural_name", "plural_name_short",
    "squad_min_play", "squad_max_play"
]

dim_position = df_positions[position_cols]
```

```python
gameweek_cols = [
    "id", "name", "deadline_time", "deadline_time_epoch"
]

dim_gameweek = df_gameweeks[gameweek_cols]
```

### Rename

```python
dim_team = dim_team.rename(columns={
    "id": "team_id",
    "name": "team_name",
    "short_name": "team_short_name"
})
```

```python
dim_player = dim_player.rename(columns={
    "id": "player_id"
})
```

```python
dim_position = dim_position.rename(columns={
    "id": "position_id"
})
```

```python
dim_player = dim_player.rename(columns={
    "id": "gameweek_id",
    "name": "gameweek_name"
})
```

### Tipe Data

```python
dim_team["team_name"] = dim_team["team_name"].astype(str)
```

### Primary Key

```python
assert dim_team["team_id"].is_unique
assert dim_team["team_id"].isna().sum() == 0
```
